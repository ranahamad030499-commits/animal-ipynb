```python
class Animal:
    def speak(self):
        return "جانور آواز نکال رہا ہے"

    def eat(self):
        return "جانور کھانا کھا رہا ہے"

    def walk(self):
        return "جانور چل رہا ہے"


class Dog(Animal):
    def speak(self):
        return "کتا: بھوں بھوں!"

    def guard(self):
        return "کتا پہرہ دے رہا ہے"


class Cat(Animal):
    def speak(self):
        return "بلی: میاؤں!"

    def sleep(self):
        return "بلی سو رہی ہے"


# Objects
dog = Dog()
cat = Cat()

print(dog.speak())
print(dog.eat())
print(dog.guard())

print(cat.speak())
print(cat.walk())
print(cat.sleep())
```

```python
class Animal:
    def __init__(self, name, sound, speed):
        self.name = name
        self.sound = sound
        self.speed = speed

    def speak(self):
        return f"{self.name} بولتا ہے: {self.sound}"

    def run(self):
        return f"{self.name} {self.speed} km/h کی رفتار سے دوڑ رہا ہے"


class Dog(Animal):
    def __init__(self):
        super().__init__("کتا", "بھوں بھوں", 40)

    def guard(self):
        return "کتا گھر کی حفاظت کر رہا ہے"


class Cat(Animal):
    def __init__(self):
        super().__init__("بلی", "میاؤں", 30)

    def jump(self):
        return "بلی زور سے چھلانگ لگا رہی ہے"


class Horse(Animal):
    def __init__(self):
        super().__init__("گھوڑا", "ہنہناہٹ", 60)


# ---- User Choice ----
print("جانور منتخب کریں:")
print("1. کتا")
print("2. بلی")
print("3. گھوڑا")

choice = input("اپنا انتخاب نمبر میں لکھیں: ")

if choice == "1":
    animal = Dog()
    print(animal.speak())
    print(animal.run())
    print(animal.guard())

elif choice == "2":
    animal = Cat()
    print(animal.speak())
    print(animal.run())
    print(animal.jump())

elif choice == "3":
    animal = Horse()
    print(animal.speak())
    print(animal.run())

else:
    print("غلط انتخاب ❌")
```
